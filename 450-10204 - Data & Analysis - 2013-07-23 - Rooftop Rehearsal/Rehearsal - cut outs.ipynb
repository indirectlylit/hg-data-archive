```python
WINDOW = 40
def getWindow(d, end):
    return utils.getDataRange(d, end-WINDOW, end)

lowVoltage = getWindow(data_2, 395)

bad = []
bad.append(getWindow(data_1, 349)) #0 - cut-out
bad.append(getWindow(data_2, 673)) #2 - cut-out
bad.append(getWindow(data_2, 943)) #3 - cut-out

good = []
good.append(getWindow(data_1, 525)) #4 - OK
good.append(getWindow(data_1, 570)) #5 - OK
good.append(getWindow(data_2, 1600)) #6 - OK
good.append(getWindow(data_2, 1750)) #6 - OK
good.append(getWindow(data_2, 1900)) #7 - OK
```

```python
def plotData(dataSet):
    time =            utils.arrayFromDictList(dataSet, 'offset')
    voltage =         utils.arrayFromDictList(dataSet, 'data', 'v')
    powerGenerated =  utils.arrayFromDictList(dataSet, 'data', 'p1')
    powerConsumed =   utils.arrayFromDictList(dataSet, 'data', 'p2')
    fig, axes = plt.subplots(nrows=2, ncols=1)
    fig.set_figheight(10)
    fig.set_figwidth(14)
    axes[0].set_title("Voltage")
    axes[0].set_ylabel("voltage")
    axes[0].set_ylim(23.5, 28.5)
    axes[0].set_xlabel("time (s)")
    axes[0].plot(time, voltage, color=utils.BROWN)
    axes[1].set_title("Power")
    axes[1].set_ylabel("watts")
    axes[1].set_ylim(0, 2000)
    axes[1].set_xlabel("time (s)")
    axes[1].plot(time, powerConsumed, color=utils.RED)
    axes[1].plot(time, powerGenerated, color=utils.BLUE)
    fig.tight_layout()
```

### Good vs Bad

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(14)

def plotData(dataset, voltageAxes, powerAxes, scenario):
    time =            utils.arrayFromDictList(dataSet, 'offset')
    voltage =         utils.arrayFromDictList(dataSet, 'data', 'v')
    powerGenerated =  utils.arrayFromDictList(dataSet, 'data', 'p1')
    powerConsumed =   utils.arrayFromDictList(dataSet, 'data', 'p2')
    voltageAxes.set_title("Voltage " + scenario)
    voltageAxes.set_ylabel("voltage")
    voltageAxes.set_ylim(23.5, 28.5)
    voltageAxes.set_xlabel("time (s)")
    voltageAxes.plot(time, voltage, color=utils.BROWN)
    powerAxes.set_title("Power Generated " + scenario)
    powerAxes.set_ylabel("watts")
    powerAxes.set_ylim(0, 1600)
    powerAxes.set_xlabel("time (s)")
    powerAxes.plot(time, powerConsumed, color=utils.RED)
    powerAxes.plot(time, powerGenerated, color=utils.BLUE)
    
for dataSet in good:
    plotData(dataSet, axes[0][0], axes[1][0], "(OK)")

for dataSet in bad:
    plotData(dataSet, axes[0][1], axes[1][1], "(Cut-out)")
    
fig.tight_layout()
```
