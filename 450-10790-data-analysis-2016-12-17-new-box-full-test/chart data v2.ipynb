```python
import bokeh
```

```python
bokeh.charts.defaults.width = 900
bokeh.charts.defaults.height = 300
```

```python
bokeh.plotting.output_notebook()
```

```python
capData = df[df.kind == 'caps'].dropna(axis=1, how='all')
acData = df[df.kind == 'acnet'].dropna(axis=1, how='all')
invData = df[df.kind == 'inv'].dropna(axis=1, how='all')
```

```python
line = bokeh.charts.TimeSeries(capData.v)
line.y_range.start = 20
line.y_range.end = 30
wheel = bokeh.models.WheelZoomTool(dimensions='width')
line.tools = [wheel,]
t = bokeh.plotting.show(line, notebook_handle=True)
```

```python
bokeh.io.push_notebook(handle=t)
```
