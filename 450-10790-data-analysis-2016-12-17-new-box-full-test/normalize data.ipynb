```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from datetime import timedelta
%matplotlib inline
sns.set(style="darkgrid")
plt.rcParams['figure.figsize'] = (12, 3)
pd.set_option('display.max_columns', 200)
# pd.set_option('display.max_rows', 100)
```

```python
all_data = pd.read_pickle('all_data.pkl')
```

```python
# convert bools to ints
for c in all_data.columns:
    if type(all_data[c].dropna()[0]) == bool:
        all_data[c] = (all_data[c] * 1.0).astype(float)
```

```python
cutoutTimestamp = pd.to_datetime("2016-12-17 20:06:09.906000")
cutoutDelta = timedelta(minutes=1, seconds=47)
videoLength = timedelta(minutes=2, seconds=0)

start = cutoutTimestamp - cutoutDelta
end = start + videoLength

data = all_data[start:end]
```

```python
groups = data.groupby(['kind', 'uid'])
```

```python
period_num = 1000
period_rule = '{}ms'.format(period_num)
resampled = groups.resample(rule=period_rule, how='mean', fill_method='ffill')
resampled.index.set_levels(resampled.index.levels[2] - resampled.index.levels[2][0], level=2, inplace=True)
resampled
```
