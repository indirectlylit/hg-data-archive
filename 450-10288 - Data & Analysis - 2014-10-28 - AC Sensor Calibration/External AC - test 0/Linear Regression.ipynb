```python
%load_ext oct2py.ipython
from oct2py import octave
import logging
octave.logger.setLevel(logging.WARN)
```

## Load data

```python
# c: list of current sensor values (in amps)
# v: list of voltage sensor values (in volts)
# p: corresponding c and v sensor values multipled by each other
# y: multimeter current and value readings multipled by each other

c, v, model_1, actual = octave.loadData()
```

## Model 1

A naive estimate of the power is to simply multiply the current and voltage values received from the sensor. This is done above and stored as `model_1`.

    predicted power = c*v

We can subtract these estimates from the values reported by the multimeters to see how the error varies with different power levels. From the graph below, it's clear that __Model 1__ is off by a non-linear factor:

```python
plt.title('Error in Estimated Power')
plt.xlabel('Actual Measured Power (W)')
plt.ylabel('Estimate Error (W)')
plt.plot(actual, np.zeros((len(actual), 1)), label='No Error')
plt.plot(actual, actual-model_1, label='Model 1')
plt.legend()
plt.show()
```

We can also get an overall sense for how far off the model is, across the entire range. The root-mean-squared error can be calculated relative to the multimeter measurements:

```python
octave.calcError(model_1, actual)
```

## Other Models

There are many other possible models that use different combinations of the basic inputs `c` and `v`. For example, the model below only uses `c` as the polynomial term rather than `c*v`, and it works basically as well. This implies that the non-linearity is in the current sensor and not the voltage sensor.

```python
%%octave -o model_3

p = c .* v;
m = length(actual);
X = [ones(m, 1) c.^2 p];
theta = normalEqn(X, actual);
model_3 = X * theta;
```

```python
plt.title('Error in Estimated Power')
plt.xlabel('Actual Measured Power (W)')
plt.ylabel('Estimate Error (W)')
plt.plot(actual, np.zeros((len(actual), 1)), label='No Error')
plt.plot(actual, actual-model_1, label='Model 1')
plt.plot(actual, actual-model_2, label='Model 2')
plt.plot(actual, actual-model_3, label='Model 3')
plt.legend()
plt.show()
```

```python
octave.calcError(model_3, actual)
```
