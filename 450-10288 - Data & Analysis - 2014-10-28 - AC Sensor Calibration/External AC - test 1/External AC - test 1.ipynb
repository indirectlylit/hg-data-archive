# External AC Sensor Calibration

```python
from oct2py import octave as oc
import logging
oc.addpath('m-files')
oc.logger.setLevel(logging.WARN)
```

## Load data

Below, we load the data into $6 \times M$ matrices, where $M$ is the number of data points and $6$ is the number of attributes for each data point. We also concatenate everything to a single data set so we can look at everything holistically.

```python
# data from text
V = 0
C_1 = 1
C_2 = 2
C_3 = 3
V_MEAS = 4
C_MEAS = 5

sensor_1 = np.loadtxt('data/sensor_1.tsv', skiprows=1)
sensor_2 = np.loadtxt('data/sensor_2.tsv', skiprows=1)
sensor_3 = np.loadtxt('data/sensor_3.tsv', skiprows=1)

# concatenate all data, discarding the ostensibly '0' columns for each data set.
V_ALL = 0
C_ALL = 1
V_ALL_MEAS = 2
C_ALL_MEAS = 3
all_sensors = oc.concatenate(sensor_1, sensor_2, sensor_3)
```

## Initial Look

We can get a quick sense of how the measured values compare to the sensor values:

```python
m = size(all_sensors, 1) # number of samples

plt.subplot(1, 2, 1)
plt.title('Voltage')
plt.xlabel('Measured Voltage (V)')
plt.ylabel('Sensor Reading (bits)')
plt.plot(sensor_1[:,V_MEAS], sensor_1[:,V], '.', label="Sensor 1", markersize=5)
plt.plot(sensor_2[:,V_MEAS], sensor_2[:,V], '.', label="Sensor 2", markersize=5)
plt.plot(sensor_3[:,V_MEAS], sensor_3[:,V], '.', label="Sensor 3", markersize=5)
plt.legend(loc="upper left")

plt.subplot(1, 2, 2)
plt.title('Current')
plt.xlabel('Measured Current (A)')
plt.ylabel('Sensor Reading (bits)')
plt.plot(sensor_1[:,C_MEAS], sensor_1[:,C_1], '.', label="Sensor 1", markersize=5)
plt.plot(sensor_2[:,C_MEAS], sensor_2[:,C_2], '.', label="Sensor 2", markersize=5)
plt.plot(sensor_3[:,C_MEAS], sensor_3[:,C_3], '.', label="Sensor 3", markersize=5)
plt.legend(loc="upper left")

plt.show()
```

This shows us a few things:

Voltage readings:

Current readings:

* not linearly related - there is a curve to the sensor data (perhaps second-order polynomial)
* relatively low random noise
* different behavior across the three sensors

## Modeling Overview

The purpose of a model is to have a way of estimating some target quantity based on sensor data features alone. A model takes the general form:

$$
h(X) = \theta X \\
= \theta_1 x_1 + \theta_2 x_2 + \dots + \theta_n x_n
$$

Where:

* $X$ is a vector of one or more features directly related to the sensor data
* $\theta$ is a vector of calibration parameters
* $h(X)$ is the predicted value ("hypothesis") of the target quantity based on the the calibration parameters and the input sensor data. Another way of saying this is: $h$ is a function parameterized by $\theta$, which takes some input $x$ and outputs a prediction $h(x)$

Modeling is a multi-step process:

1. Collect sensor data alongside known ("ground truth") values of the target quantity to use for training
2. Choose a structure for the model based on manual inspection of the data and the ground truth values
3. Find parameters ($\theta$) which fit the training data to the chosen model structure
4. Use the parameterized model to predict values of the target quantity based on new sensor data

## Voltage Model

Since the actual voltage values appear to be linearly related to the sensor readings, we can expect the model to take the form:

$$
h(v) = \theta_1 + \theta_2 v
$$

where $\theta_1$ and $\theta_2$ are constants, and $v$ is the sensor reading in bits.

We can find optimal $\theta$ values using the known voltage values from the multimeter $y$ and finding values for theta which, on average, minimize the error in $h(x)$ compared to $y$:

```python
sensorVals = all_sensors[:, V_ALL]
voltageReadings = all_sensors[:, V_ALL_MEAS].T
X_v, theta_v, h_v = oc.optimizeVoltageModel(sensorVals, voltageReadings)
print("theta_1: {}\ntheta_2: {}".format(float(theta_v[0]), float(theta_v[1])))
```

Given these values, we can compare the predicted voltage values with the measured values:

```python
rmsErr_v, errors_v = oc.calcError(h_v.T, voltageReadings)

plt.title('Voltage Error')
plt.xlabel('Measured Voltage (V)')
plt.ylabel('Error in predicted voltage (V)')
plt.plot(voltageReadings, oc.zeros((len(voltageReadings), 1)), '-', markersize=5)
plt.plot(voltageReadings, errors_v, '.', markersize=5)
plt.show()

print("The average root-mean-squared error with this model and sample data is {0:.2f} volts".format(float(rmsErr_v)))
```

## Current Model

The initial look at the data appeared to imply a non-linear relationship between the actual current value as measured by the multimeters and the sensor readings. Therefore we might expect a model to look like:

$$
h(c) = \theta_1 + \theta_2 c + \theta_3 c^2
$$

where the $\theta$ values are constants and $c$ is the sensor reading in bits.

Since the three sensors clearly had different curves, we need to compute each set of $\theta$ parameters individually.

```python
X_c_1, theta_c_1, h_c_1 = oc.optimizeCurrentModel(sensor_1[:,C_1], sensor_1[:, C_MEAS].T)
print("Sensor 1")
print("  theta_1: {}".format(float(theta_c_1[0])))
print("  theta_2: {}".format(float(theta_c_1[1])))
print("  theta_3: {}".format(float(theta_c_1[2])))

X_c_2, theta_c_2, h_c_2 = oc.optimizeCurrentModel(sensor_2[:,C_2], sensor_2[:, C_MEAS].T)
print("Sensor 2")
print("  theta_1: {}".format(float(theta_c_2[0])))
print("  theta_2: {}".format(float(theta_c_2[1])))
print("  theta_3: {}".format(float(theta_c_2[2])))

X_c_3, theta_c_3, h_c_3 = oc.optimizeCurrentModel(sensor_3[:,C_3], sensor_3[:, C_MEAS].T)
print("Sensor 3")
print("  theta_1: {}".format(float(theta_c_3[0])))
print("  theta_2: {}".format(float(theta_c_3[1])))
print("  theta_3: {}".format(float(theta_c_3[2])))
```

And looking at the errors:

```python
c = sensor_1[:, C_MEAS].T
rmsErr_c_1, errors_c_1 = oc.calcError(h_c_1.T, c)

plt.subplot(1, 3, 1)
plt.title('Sensor 1')
plt.ylabel('Error in predicted current (A)')
plt.plot(c, oc.zeros((len(c), 1)), '-', markersize=5)
plt.plot(c, errors_c_1, '.', markersize=5)
plt.ylim(-0.16, 0.16)

c = sensor_2[:, C_MEAS].T
rmsErr_c_2, errors_c_2 = oc.calcError(h_c_2.T, c)

plt.subplot(1, 3, 2)
plt.title('Sensor 2')
plt.xlabel('Measured Current (A)')
plt.plot(c, oc.zeros((len(c), 1)), '-', markersize=5)
plt.plot(c, errors_c_2, '.', markersize=5)
plt.ylim(-0.16, 0.16)

c = sensor_3[:, C_MEAS].T
rmsErr_c_3, errors_c_3 = oc.calcError(h_c_3.T, c)

plt.subplot(1, 3, 3)
plt.title('Sensor 3')
plt.plot(c, oc.zeros((len(c), 1)), '-', markersize=5)
plt.plot(c, errors_c_3, '.', markersize=5)
plt.ylim(-0.16, 0.16)

plt.show()

print("Avg RMS errors are {}, {}, {} amps, respectively".format(rmsErr_c_1, rmsErr_c_2, rmsErr_c_3))
```

Some observations:

* While the model seems to work well for sensors 2 and 3, it isn't great for sensor 1. Looking at the initial data, sensor 1 does seem to look different, and has an inflection point around 5 amps.
* Sensors 2 and 3 show data points with larger errors at the lower current levels, which may indicate that the sensor has some floor value and values below this should not be trusted.
